# Creating an agent class
### that has the propeties of looking the neighbours & has its what type and the position 

```python
import random
```

```python
class Agent:
    def __init__(self,agent_type,position,tolerance):
        self.agent_type = agent_type
        self.position = position   
        self.tolerance = tolerance
        
    # check the neighbours and its type
    def is_happy(self,neighbours):
        same_type = 0
        total_neighbours = 0

        for n in neighbours:
            if n == ".":
                continue
                
                #count the total number of neighbours
            total_neighbours += 1
            
            if n == self.agent_type:
                same_type += 1

            if n == self.agent_type:
                same_type += 1
                
        if total_neighbours == 0:
            return True
        trushold = same_type / total_neighbours
        
        return trushold >= self.tolerance
            
    def display(self):
        print("Agent type",self.agent_type)
        print("Agent Position",self.position)
    def check_neighbours(self,grid):
        # the characteristics of the neighbours to be defined later
        neighors =[]
        row, col = self.position
        
        #formation o the grid
        for dr in [-1,0,1]:
            for dc in [-1,0,1]:
                if dr == 0 and dc == 0:
                    continue
                    
                #telling the program that we will find  the neighours in these all position 
                new_row = row + dr
                new_col = row + dc
                
                # check the overall grid and append the value of the neighbour
                if 0 <= new_row < len(grid) and new_col < len(grid[0]):
                    neighbors.append(grid[new_row][new_col])
        return neighbours

        #Movement function
        def move(self,grid,empty_cells):
            new_position = random.choice(empty_cells)
            old_row,old_col = self.position
            grid[old_row][old_col] = "."

            new_row, new_col = new_position
            grid[new_row][new_col] = self.agent_type

            self.position = new_position 
                
        
```

```python
agent1 = Agent("Red",(2,3),0.5)
agent1.display()
```

```python
agent2 = Agent("Blue",(4,1),0.5)
agent2.display()
```
